```python
class LanguageIDDataset(Dataset):
    def __init__(self, df, feature_extractor, target_len_sec=5):
        self.df = df
        self.feature_extractor = feature_extractor
        self.target_len = int(16000 * target_len_sec)  # e.g., 5 seconds

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            audio_path = row["file_path"]
            label = row["language_code"]

            waveform, sr = torchaudio.load(audio_path)
            waveform = waveform.mean(dim=0)  # convert to mono

            if sr != 16000:
                waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)(waveform)

            total_len = waveform.shape[0]
            if total_len <= self.target_len:
                pad_len = self.target_len - total_len
                waveform = torch.nn.functional.pad(waveform, (0, pad_len))
                segments = [waveform]
            else:
                segments = []
                start = 0
                while start < total_len:
                    end = min(start + self.target_len, total_len)
                    chunk = waveform[start:end]
                    if chunk.shape[0] < self.target_len:
                        pad_len = self.target_len - chunk.shape[0]
                        chunk = torch.nn.functional.pad(chunk, (0, pad_len))
                    segments.append(chunk)
                    start += self.target_len

            # Return first chunk only
            inputs = self.feature_extractor(
                segments[0].numpy(), sampling_rate=16000, return_tensors="pt", padding=False
            )
            input_values = inputs["input_values"].squeeze(0)
            return input_values, torch.tensor(label, dtype=torch.long)

        except Exception as e:
            print(f"[Dataset] Skipping audio {audio_path} due to error: {e}")
            return None
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from transformers import HubertModel
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import gc

# Set up directories
os.makedirs("results_fine_tune_extended_extended", exist_ok=True)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define Model
class HubertClassifier(nn.Module):
    def __init__(self, num_labels):
        super(HubertClassifier, self).__init__()
        self.hubert = HubertModel.from_pretrained("facebook/hubert-large-ls960-ft")
        self.classifier = nn.Sequential(
            nn.Linear(self.hubert.config.hidden_size, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_labels)
        )

    def forward(self, input_values):
        outputs = self.hubert(input_values).last_hidden_state
        pooled = outputs.mean(dim=1)
        return self.classifier(pooled)

# Instantiate
num_classes = df['language_code'].nunique()
model = HubertClassifier(num_labels=num_classes).to(device)
criterion = nn.CrossEntropyLoss()

# Training Stats
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

# --------------- Phase 1: Feature Extraction -----------------
feature_extract_epochs = 10
for param in model.hubert.parameters():
    param.requires_grad = False
optimizer = optim.Adam(model.classifier.parameters(), lr=1e-3)

for epoch in range(feature_extract_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 200: break
        try:
            input_values, labels = batch
            input_values, labels = input_values.to(device), labels.to(device)
        except Exception as e:
            print(f"[Feature Extract] Skipping batch {bcnt}: {e}")
            continue

        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()

    acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(acc)
    print(f"[Feature Extract] Epoch {epoch+1}/{feature_extract_epochs} - Loss: {total_loss:.4f}, Acc: {acc*100:.2f}%")

# Save feature extraction logs
with open("results_fine_tune_extended/feature_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

# --------------- Phase 2: Fine-Tuning -----------------
fine_tune_epochs = 7
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

for param in model.hubert.parameters():
    param.requires_grad = True
optimizer = optim.Adam(model.parameters(), lr=1e-4)

for epoch in range(fine_tune_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 300: break
        try:
            input_values, labels = batch
            input_values, labels = input_values.to(device), labels.to(device)
        except Exception as e:
            print(f"[Fine-tune] Skipping batch {bcnt}: {e}")
            continue

        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()

    train_acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(train_acc)

    print(f"[Fine-tune] Epoch {epoch+1}/{fine_tune_epochs} - Loss: {total_loss:.4f}, Acc: {train_acc*100:.2f}%")

    # -------- Validation --------
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for vcnt, batch in enumerate(val_loader):
            if vcnt >= 200: break
            try:
                input_values, labels = batch
                input_values, labels = input_values.to(device), labels.to(device)
            except Exception as e:
                print(f"[Validation] Skipping batch {vcnt}: {e}")
                continue

            outputs = model(input_values)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            preds = outputs.argmax(dim=1)
            val_preds.extend(preds.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())

            del input_values, labels, outputs, loss, preds
            torch.cuda.empty_cache()
            gc.collect()

    val_acc = accuracy_score(val_labels, val_preds)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)

    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc*100:.2f}%")

    # Save checkpoints
    if (epoch+1) in [5, 10, 11,12,14,16]:
        torch.save(model.state_dict(), f"results_fine_tune_extended/hubert.pt")
        print("MOdel saved at epoch",epoch+1)

# Save fine-tune logs
with open("results_fine_tune_extended/fine_tune_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

with open("results_fine_tune_extended/fine_tune_val_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(val_losses, val_accuracies), 1):
        f.write(f"Epoch {e}: Val Loss={l:.4f}, Val Accuracy={a*100:.2f}%\n")

# --------- Plot Results ----------
epochs = list(range(1, fine_tune_epochs + 1))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label="Train Loss", color='blue')
plt.plot(epochs, val_losses, label="Val Loss", color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Over Epochs")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs, [acc * 100 for acc in train_accuracies], label="Train Acc", color='green')
plt.plot(epochs, [acc * 100 for acc in val_accuracies], label="Val Acc", color='red')
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy Over Epochs")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("results_fine_tune_extended/training_curves.png")
plt.close()
```

```python
import os
import gc
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchaudio
from transformers import HubertModel, Wav2Vec2FeatureExtractor
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Set up directories
os.makedirs("results_fine_tune_extended", exist_ok=True)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ------------- Dataset and Collate Function ------------- #
class LanguageIDDataset(Dataset):
    def __init__(self, df, feature_extractor, target_len_sec=5):
        self.df = df
        self.feature_extractor = feature_extractor
        self.target_len = int(16000 * target_len_sec)  # e.g., 5 seconds

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            audio_path = row["file_path"]
            label = row["language_code"]

            waveform, sr = torchaudio.load(audio_path)
            waveform = waveform.mean(dim=0)  # convert to mono

            if sr != 16000:
                waveform = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)(waveform)

            total_len = waveform.shape[0]
            if total_len <= self.target_len:
                pad_len = self.target_len - total_len
                waveform = torch.nn.functional.pad(waveform, (0, pad_len))
                segments = [waveform]
            else:
                segments = []
                start = 0
                while start < total_len:
                    end = min(start + self.target_len, total_len)
                    chunk = waveform[start:end]
                    if chunk.shape[0] < self.target_len:
                        pad_len = self.target_len - chunk.shape[0]
                        chunk = torch.nn.functional.pad(chunk, (0, pad_len))
                    segments.append(chunk)
                    start += self.target_len

            inputs = self.feature_extractor(
                segments[0].numpy(), sampling_rate=16000, return_tensors="pt", padding=False
            )
            input_values = inputs["input_values"].squeeze(0)
            return input_values, torch.tensor(label, dtype=torch.long)

        except Exception as e:
            print(f"[Dataset] Skipping audio {self.df.iloc[idx]['file_path']} due to error: {e}")
            return None

def language_id_collate_fn(batch):
    batch = [item for item in batch if item is not None]
    if len(batch) == 0:
        return None, None
    input_values = torch.stack([b[0] for b in batch])
    labels = torch.tensor([b[1] for b in batch], dtype=torch.long)
    return input_values, labels

# ------------- Model ------------- #
class HubertClassifier(nn.Module):
    def __init__(self, num_labels):
        super(HubertClassifier, self).__init__()
        self.hubert = HubertModel.from_pretrained("facebook/hubert-large-ls960-ft")
        self.classifier = nn.Sequential(
            nn.Linear(self.hubert.config.hidden_size, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_labels)
        )

    def forward(self, input_values):
        outputs = self.hubert(input_values).last_hidden_state
        pooled = outputs.mean(dim=1)
        return self.classifier(pooled)

# ------------ Setup ------------ #
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/hubert-large-ls960-ft")
train_dataset = LanguageIDDataset(train_df, feature_extractor)
val_dataset = LanguageIDDataset(val_df, feature_extractor)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=language_id_collate_fn)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, collate_fn=language_id_collate_fn)

num_classes = train_df['language_code'].nunique()
model = HubertClassifier(num_labels=num_classes).to(device)
criterion = nn.CrossEntropyLoss()

# ------------ Phase 1: Feature Extraction ------------ #
train_losses, train_accuracies = [], []
feature_extract_epochs = 10
for param in model.hubert.parameters():
    param.requires_grad = False
optimizer = optim.Adam(model.classifier.parameters(), lr=1e-3)

for epoch in range(feature_extract_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 200: break
        input_values, labels = batch
        if input_values is None or labels is None:
            print(f"[Feature Extract] Skipping empty batch {bcnt}")
            continue

        input_values, labels = input_values.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()

    acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(acc)
    print(f"[Feature Extract] Epoch {epoch+1}/{feature_extract_epochs} - Loss: {total_loss:.4f}, Acc: {acc*100:.2f}%")

# Save feature extraction logs
with open("results_fine_tune_extended/feature_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

# ------------ Phase 2: Fine-Tuning ------------ #
fine_tune_epochs = 7
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

for param in model.hubert.parameters():
    param.requires_grad = True
optimizer = optim.Adam(model.parameters(), lr=1e-4)

for epoch in range(fine_tune_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 300: break
        input_values, labels = batch
        if input_values is None or labels is None:
            print(f"[Fine-tune] Skipping empty batch {bcnt}")
            continue

        input_values, labels = input_values.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()

    train_acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(train_acc)

    print(f"[Fine-tune] Epoch {epoch+1}/{fine_tune_epochs} - Loss: {total_loss:.4f}, Acc: {train_acc*100:.2f}%")

    # Validation
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for vcnt, batch in enumerate(val_loader):
            if vcnt >= 200: break
            input_values, labels = batch
            if input_values is None or labels is None:
                print(f"[Validation] Skipping empty batch {vcnt}")
                continue

            input_values, labels = input_values.to(device), labels.to(device)
            outputs = model(input_values)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            preds = outputs.argmax(dim=1)
            val_preds.extend(preds.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())

            del input_values, labels, outputs, loss, preds
            torch.cuda.empty_cache()
            gc.collect()

    val_acc = accuracy_score(val_labels, val_preds)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)
    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc*100:.2f}%")

    # Checkpoint (optional)
    if (epoch+1) in [5, 7]:
        torch.save(model.state_dict(), f"results_fine_tune_extended/hubert_epoch_{epoch+1}.pt")
        print("Model saved at epoch", epoch+1)

# Save fine-tune logs
with open("results_fine_tune_extended/fine_tune_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

with open("results_fine_tune_extended/fine_tune_val_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(val_losses, val_accuracies), 1):
        f.write(f"Epoch {e}: Val Loss={l:.4f}, Val Accuracy={a*100:.2f}%\n")

# Plot
epochs = list(range(1, fine_tune_epochs + 1))
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label="Train Loss", color='blue')
plt.plot(epochs, val_losses, label="Val Loss", color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Over Epochs")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs, [acc * 100 for acc in train_accuracies], label="Train Acc", color='green')
plt.plot(epochs, [acc * 100 for acc in val_accuracies], label="Val Acc", color='red')
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy Over Epochs")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("results_fine_tune_extended/training_curves.png")
plt.close()
```
