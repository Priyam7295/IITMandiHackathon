```python
from torch.utils.data import Dataset
import torch
import torchaudio

class LanguageIDDataset(Dataset):
    def __init__(self, df, feature_extractor, target_len_sec=5, num_segments=2):
        self.df = df
        self.feature_extractor = feature_extractor
        self.target_len = int(16000 * target_len_sec)  
        self.total_len = self.target_len * num_segments  
        self.num_segments = num_segments

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        try:
            row = self.df.iloc[idx]
            audio_path = row["file_path"]
            label = row["language_code"]

            waveform, sr = torchaudio.load(audio_path)
            waveform = waveform.mean(dim=0) 
            
            if sr != 16000:
                resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
                waveform = resampler(waveform)

            total_len = waveform.shape[0]
            segments = []
            start = 0

            # Collect up to num_segments of target_len each
            while start < total_len and len(segments) < self.num_segments:
                end = min(start + self.target_len, total_len)
                chunk = waveform[start:end]
                if chunk.shape[0] < self.target_len:
                    pad_len = self.target_len - chunk.shape[0]
                    chunk = torch.nn.functional.pad(chunk, (0, pad_len))
                segments.append(chunk)
                start += self.target_len

            # Pad with silence (zeros) if less than 3 chunks
            while len(segments) < self.num_segments:
                segments.append(torch.zeros(self.target_len))

            # Concatenate into one long waveform (e.g., 15 seconds)
            combined_waveform = torch.cat(segments)

            inputs = self.feature_extractor(
                combined_waveform.numpy(),
                sampling_rate=16000,
                return_tensors="pt",
                padding=False
            )
            input_values = inputs["input_values"].squeeze(0)
            return input_values, torch.tensor(label, dtype=torch.long)

        except Exception as e:
            print(f"[Dataset] Skipping audio {audio_path} due to error: {e}")
            return None
```

```python
import pandas as pd
import torch
from torch.utils.data import DataLoader
from transformers import Wav2Vec2FeatureExtractor
from sklearn.model_selection import train_test_split
from torch.utils.data.dataloader import default_collate

# Load CSV
df = pd.read_csv("/home/dl/Downloads/Shyam copy/label/ground_truth.csv")

# Shuffle and split (80:10:10)
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, shuffle=True)

# Feature extractor
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-large-xlsr-53")

# Custom Dataset (you must have this defined properly)
# Assuming your LanguageIDDataset is implemented to return None if it fails to load a sample
train_dataset = LanguageIDDataset(train_df, feature_extractor)
val_dataset = LanguageIDDataset(val_df, feature_extractor)
test_dataset = LanguageIDDataset(test_df, feature_extractor)

# Custom collate function that skips None samples
def collate_skip_none(batch):
    batch = [item for item in batch if item is not None]
    if len(batch) == 0:
        return None
    return default_collate(batch)

# DataLoaders with collate_fn
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, collate_fn=collate_skip_none)
val_loader = DataLoader(val_dataset, batch_size=16, collate_fn=collate_skip_none)
test_loader = DataLoader(test_dataset, batch_size=16, collate_fn=collate_skip_none)

# Visualize one batch
for batch in train_loader:
    if batch is None:
        continue
    input_values, labels = batch
    print("Input shape:", input_values.shape)
    print("Labels:", labels)
    break
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from transformers import HubertModel
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import gc

# For reproducibility and safety
torch.backends.cudnn.benchmark = False
torch.backends.cudnn.deterministic = True

# Set up directories
os.makedirs("results_fine_tune_fb", exist_ok=True)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define Model
import torch.nn.functional as F

class HubertClassifier(nn.Module):
    def __init__(self, num_labels):
        super(HubertClassifier, self).__init__()
        self.hubert = HubertModel.from_pretrained("facebook/hubert-base-ls960")

        self.classifier = nn.Sequential(
            nn.Linear(self.hubert.config.hidden_size, 512),
            nn.LayerNorm(512),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(256, 128),
            nn.LayerNorm(128),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(128, num_labels)
        )

    def forward(self, input_values):
        with torch.amp.autocast(device_type='cuda'):  # updated mixed precision
            outputs = self.hubert(input_values).last_hidden_state
            pooled = outputs.mean(dim=1)
            return self.classifier(pooled)



# Instantiate
num_classes = df['language_code'].nunique()
model = HubertClassifier(num_labels=num_classes).to(device)
criterion = nn.CrossEntropyLoss()

train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

# --------------- Phase 1: Feature Extraction -----------------
feature_extract_epochs = 7
for param in model.hubert.parameters():
    param.requires_grad = False
optimizer = optim.Adam(model.classifier.parameters(), lr=1e-3)

for epoch in range(feature_extract_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 200: break
        try:
            input_values, labels = batch
            input_values, labels = input_values.to(device), labels.to(device)
        except Exception as e:
            print(f"[Feature Extract] Skipping batch {bcnt}: {e}")
            continue

        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()
        torch.cuda.ipc_collect()

    acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(acc)

    torch.cuda.empty_cache()
    gc.collect()
    torch.cuda.ipc_collect()

    print(f"[Feature Extract] Epoch {epoch+1}/{feature_extract_epochs} - Loss: {total_loss:.4f}, Acc: {acc*100:.2f}%")

    if (epoch+1) in [4,6,9]:
        torch.save(model.state_dict(), f"results_fine_tune_fb/hubert.pt")
        print("Model saved at epoch", epoch+1)

with open("results_fine_tune_fb/feature_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

# --------------- Phase 2: Fine-Tuning -----------------
fine_tune_epochs = 3
train_losses, train_accuracies = [], []
val_losses, val_accuracies = [], []

for param in model.hubert.parameters():
    param.requires_grad = True
optimizer = optim.Adam(model.parameters(), lr=1e-4)

for epoch in range(fine_tune_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for bcnt, batch in enumerate(train_loader):
        if bcnt >= 200: break
        try:
            input_values, labels = batch
            input_values, labels = input_values.to(device), labels.to(device)
        except Exception as e:
            print(f"[Fine-tune] Skipping batch {bcnt}: {e}")
            continue

        optimizer.zero_grad()
        outputs = model(input_values)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

        del input_values, labels, outputs, loss, preds
        torch.cuda.empty_cache()
        gc.collect()
        torch.cuda.ipc_collect()

    train_acc = accuracy_score(all_labels, all_preds)
    train_losses.append(total_loss)
    train_accuracies.append(train_acc)

    torch.cuda.empty_cache()
    gc.collect()
    torch.cuda.ipc_collect()

    print(f"[Fine-tune] Epoch {epoch+1}/{fine_tune_epochs} - Loss: {total_loss:.4f}, Acc: {train_acc*100:.2f}%")

    # -------- Validation --------
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for vcnt, batch in enumerate(val_loader):
            if vcnt >= 200: break
            try:
                input_values, labels = batch
                input_values, labels = input_values.to(device), labels.to(device)
            except Exception as e:
                print(f"[Validation] Skipping batch {vcnt}: {e}")
                continue

            outputs = model(input_values)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            preds = outputs.argmax(dim=1)
            val_preds.extend(preds.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())

            del input_values, labels, outputs, loss, preds
            torch.cuda.empty_cache()
            gc.collect()
            torch.cuda.ipc_collect()

    val_acc = accuracy_score(val_labels, val_preds)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)

    torch.cuda.empty_cache()
    gc.collect()
    torch.cuda.ipc_collect()

    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc*100:.2f}%")

    if (epoch+1) in [4,6,9]:
        torch.save(model.state_dict(), f"results_fine_tune_fb/hubert.pt")
        print("Model saved at epoch", epoch+1)

with open("results_fine_tune_fb/fine_tune_train_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(train_losses, train_accuracies), 1):
        f.write(f"Epoch {e}: Loss={l:.4f}, Accuracy={a*100:.2f}%\n")

with open("results_fine_tune_fb/fine_tune_val_log.txt", "w") as f:
    for e, (l, a) in enumerate(zip(val_losses, val_accuracies), 1):
        f.write(f"Epoch {e}: Val Loss={l:.4f}, Val Accuracy={a*100:.2f}%\n")

# --------- Plot Results ----------
epochs = list(range(1, fine_tune_epochs + 1))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label="Train Loss", color='blue')
plt.plot(epochs, val_losses, label="Val Loss", color='orange')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Over Epochs")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs, [acc * 100 for acc in train_accuracies], label="Train Acc", color='green')
plt.plot(epochs, [acc * 100 for acc in val_accuracies], label="Val Acc", color='red')
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy Over Epochs")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("results_fine_tune_fb/training_curves.png")
plt.close()
```

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import torch
import gc

# ---- Test Phase (Limit to 2000 samples) ----
test_preds, test_labels = [], []
sample_count = 0
max_samples = 2000

model.eval()
with torch.no_grad():
    for tcnt, batch in enumerate(test_loader):
        try:
            input_values, labels = batch
            input_values, labels = input_values.to(device), labels.to(device)
        except Exception as e:
            print(f"[Test] Skipping batch {tcnt}: {e}")
            continue

        outputs = model(input_values)
        preds = outputs.argmax(dim=1)

        for p, l in zip(preds.cpu().numpy(), labels.cpu().numpy()):
            test_preds.append(p)
            test_labels.append(l)
            sample_count += 1
            if sample_count >= max_samples:
                break

        del input_values, labels, outputs, preds
        torch.cuda.empty_cache()
        gc.collect()
        torch.cuda.ipc_collect()

        if sample_count >= max_samples:
            break

test_preds = np.array(test_preds)
test_labels = np.array(test_labels)

# ---- Metrics ----
f1 = f1_score(test_labels, test_preds, average="weighted")
conf_mat = confusion_matrix(test_labels, test_preds)
report = classification_report(test_labels, test_preds, digits=4)

print(f"\nWeighted F1-score (first 2000 samples): {f1:.4f}\n")
print("Classification Report (first 2000 samples):\n", report)

# ---- Confusion Matrix Plot ----
plt.figure(figsize=(10, 8))
sns.heatmap(conf_mat, annot=True, fmt="d", cmap="Blues", xticklabels=label_names, yticklabels=label_names)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix (First 2000 Samples)")
plt.tight_layout()
plt.savefig("results_fine_tune_fb/confusion_matrix_2000.png")
plt.close()
```

```python
import os
import torchaudio
from tqdm import tqdm


label_map_class=["Kannada","Marathi","Punjabi","Telugu","Gujarati","Malayalam","Urdu","Tamil","Hindi","Bengali"]

def predict_real_world_audio(model, folder_path, feature_extractor, device, chunk_duration_sec=5):
    cnt=0
    model.eval()
    sample_rate = 16000
    chunk_samples = chunk_duration_sec * sample_rate

    files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.wav', '.mp3'))]
    print(f" Found {len(files)} audio files in '{folder_path}'")

    for file in tqdm(files, desc="Processing files"):
        cnt+=1
        filepath = os.path.join(folder_path, file)
        display(Audio(filename=filepath))
        try:
            waveform, sr = torchaudio.load(filepath)
            waveform = waveform.mean(dim=0)  # Mono

            if sr != sample_rate:
                resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=sample_rate)
                waveform = resampler(waveform)

            total_len = waveform.shape[0]
            num_chunks = (total_len + chunk_samples - 1) // chunk_samples

            print(f"\n File: {file} | Duration: {total_len / sample_rate:.2f}s | Chunks: {num_chunks}")
            for i in range(num_chunks):
                start = i * chunk_samples
                end = min((i + 1) * chunk_samples, total_len)
                chunk = waveform[start:end]

                # Pad if needed
                if chunk.shape[0] < chunk_samples:
                    pad_len = chunk_samples - chunk.shape[0]
                    chunk = torch.nn.functional.pad(chunk, (0, pad_len))

                inputs = feature_extractor(chunk.numpy(), sampling_rate=sample_rate, return_tensors="pt")
                input_values = inputs["input_values"].to(device)
                with torch.no_grad():
                    outputs = model(input_values)
                    rounded_outputs = [round(score.item(), 2) for score in outputs[0]]  # Round each class score
                    print(f" Model Output (rounded): {rounded_outputs}")
                    pred = torch.argmax(outputs, dim=1).item()

                print(f"   Chunk {i+1}: Predicted Language Index -> {pred} ")
                print(label_map_class[pred])
            print("---------------------------------------------------------------------------------------------\n\n\n")

        except Exception as e:
            print(f" Failed to process '{file}': {e}")
        if cnt>=9:
            break
```
